```python
'SELECT gaia.source_id, gaia.ra, gaia.ra_error, gaia.dec,gaia.dec_error, \
gaia.pmra, gaia.pmra_error, gaia.pmdec, gaia.pmdec_error, gaia.radial_velocity, \
gaia.radial_velocity_error, gaia.parallax, gaia.parallax_error, gaia.phot_g_mean_mag, \
gaia.phot_bp_mean_mag, gaia.phot_rp_mean_mag \
FROM gaiadr2.gaia_source as gaia \
WHERE gaia.parallax > 5 and \
gaia.pmra_error < .1 and gaia.pmdec_error < .1 and gaia.parallax_error < .1 and gaia.radial_velocity_error < 1 and \
gaia.pmra is not null and gaia.pmdec is not null and \
gaia.phot_bp_mean_mag is not null and gaia.phot_rp_mean_mag is not null and \
gaia.phot_g_mean_mag is not null and gaia.radial_velocity is not null'

# http://docs.astropy.org/en/stable/generated/examples/coordinates/plot_galactocentric-frame. 
# html#sphx-glr-generated-examples-coordinates-plot-galactocentric-frame-py

# https://arxiv.org/pdf/1105.6093.pdf

cols = ['x','y','z','u','v','w']

%matplotlib inline
import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d
import seaborn as sns 
import statistics as stat
import scipy.stats as st
import random as rand
from astropy.table import Table
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
import astropy.coordinates as coord
import astropy.units as u

def pc_ly(x):
    return (1/(float(x)/1000))*3.26156
def mas_deg(x):
    return x*(1/1000)/3600
def abs_mag(x,y):
    return x - (5 * (np.log10(y/10))) 
def color(x,y):
    return [y[i]-x[i] for i in range(len(x))]
def z_score(x,y,z):
    return (x-y)/z
def rnd(x):
    return round(x,2)
def rnd4(x):
    return round(x,4)
def dist(x):
    return 1/(x/1000)
# removes unit in order to operate with astropy objects 
def V(myarg):
    return (myarg / myarg.unit).value
def sixspace(data,_):
    v_sun = coord.CartesianDifferential([0,0,0]*u.km/u.s) # [11.1, 244, 7.25]
    c = coord.ICRS(ra = data['ra'][_]*u.degree, dec=data['dec'][_]*u.degree,
                distance=(data['parallax'][_]*u.mas).to(u.pc, u.parallax()),
                pm_ra_cosdec=data['pmra'][_]*u.mas/u.yr,
                pm_dec=data['pmdec'][_]*u.mas/u.yr,
                radial_velocity=data['radial_velocity'][_]*u.km/u.s)

    gc_frame = coord.Galactocentric(galcen_distance=8*u.kpc,
                                galcen_v_sun=v_sun,
                                z_sun=0*u.pc)
    gc = c.transform_to(gc_frame)
    return [V(gc.x)+8000, V(gc.y), V(gc.z), V(gc.v_x), V(gc.v_y), V(gc.v_z)]
def astroFrame(data):
    lis = []
    for i in range(len(data)):
        lis.append(sixspace(data,i))
    return pd.DataFrame(lis,columns = cols)
```

```python
def spherical(x,y,z):
    rho = np.sqrt(x**2+y**2+z**2)
    phi = np.arccos(z/rho)
    theta = np.arctan(y/x)
    lis = [rho,abs(theta),abs(phi)]
    return lis

RHO, THETA, PHI = list(np.arange(0,350,50)), list(np.mgrid[0:2*np.pi:20j]), list(np.mgrid[0:np.pi:10j])
THETA.append(2*np.pi), PHI.append(np.pi)

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Partitions')

partitions = {}
dV = 0
for r in range(len(RHO)-1):
    for t in range(len(THETA)-1):
        for p in range(len(PHI)-1): 
            ax.scatter3D(RHO[r]*np.cos(THETA[t])*np.sin(PHI[p]),\
                         RHO[r]*np.sin(THETA[t])*np.sin(PHI[p]),\
                         RHO[r]*np.cos(PHI[p]), c='k', s=7)
            iD = []
            for i in range(len(tail)):
                    coords = spherical(tail['x'][i] + 43.43, tail['y'][i] + 0.51, tail['z'][i] + 16.74)
                    if coords[0] >= RHO[r] and coords[1] >= THETA[t] and coords[2] >= PHI[p] and \
                    coords[0] <  RHO[r+1] and coords[1] < THETA[t+1] and coords[2] < PHI[p+1]:
                        iD.append(tail['id'][i])
            # partitions =  [[list of ids to retrieve pm data etc(index in FILT)],index in tails, spherical constraints...]
            partitions[dV] = [iD,i,RHO[r],np.rad2deg(RHO[r+1]),np.rad2deg(THETA[t]), \
                              np.rad2deg(THETA[t+1]),np.rad2deg(PHI[p]),np.rad2deg(PHI[p+1])]
            dV += 1
            
ax.scatter3D(tail['x']+43.43, tail['y']+0.51, tail['z']+16.74, c='r',s=1,alpha=.3)
plt.show()
```

```python
from astropy.table import Table
from astropy.visualization import astropy_mpl_style
plt.style.use(astropy_mpl_style)
import astropy.coordinates as coord
import astropy.units as u

## Extrapolating proper motions for the stars within Volume(i) in order to determine POC
master = []

for part in range(len(partitions)):
    samp = [i for i in partitions[part][0]]
    rex = [list((FILT.loc[i])) for i in samp]
    members = pd.DataFrame(rex,columns = list(FILT))
    gal_membs = astroFrame(members)
    ra0, dec0 = members['ra'], members['dec'] 
    pmra, pmdec = mas_deg(members['pmra']), mas_deg(members['pmdec'])
    stars = {}
    #plt.figure(figsize=(8,8))
    #plt.ylabel('dec (deg)')
    #plt.xlabel('ra (deg)')
    #plt.xlim([120,35])
    #plt.ylim([-7,37])
    #plt.title('Phase 1 - {} stars'.format(len(members)))
    
    OOO = len(gal_membs)
    if OOO != 0:
        for memb in range(len(members)):
            ra = []
            dec = []
            for x in range(1,5000,10):
                ra.append(ra0[memb]+(pmra[memb]*x*2000))
                dec.append(dec0[memb]+(pmdec[memb]*x*2000))
            stars[memb] = [ra,dec]
            #plt.scatter(ra, dec, s = .2, c = 'k')
        #plt.show() ------ Un hashtag to show plots for V(i)

        # Calculating POC for V(i) (dV size to be narrowed down)
        dA = {}
        step_ra = 2.5
        step_dec = 2.5
        BOX = 0
        RA = np.arange(80,120,2.5) #2.5
        DEC = np.arange(0,15,2.5) #2.5

        for R in RA:
            for D in DEC:
                appearance = 0
                for i in range(len(stars)):
                    for j in range(len(stars[i][0])):
                        if stars[i][0][j] > R and stars[i][0][j] < R + step_ra \
                        and stars[i][1][j] > D and stars[i][1][j] < D + step_dec:
                            appearance += 1
                dA[BOX] = [appearance,R,R+step_ra,D,D+step_dec]
                BOX += 1

        # POC of Core Members: [99,96] x [9,7] 
        # subjective POC: This will be narrowed down along with dV partition size to optimze its the algo sensitivity 
        # to different spatial distributions and predicted POC dispersion.  

        # [90,105] x[5,10]
        #subjpoc_ra = [85,110]
        #subjpoc_dec = [0,15]
        subjpoc_ra = [70,120]
        subjpoc_dec = [-10,25]
        boxes = [(num-1,dA[i]) for num ,i in enumerate(dA,1)]
        for i in range(len(boxes)):
            if boxes[i][1] == max([boxes[i][1] for i in range(len(boxes))]):
                POC = dA[boxes[i][0]][1:]
                POCra = np.mean([POC[0],POC[1]])
                POCdec = np.mean([POC[2],POC[3]])
                #print('POCra {} POCdec {}'.format(POCra,POCdec))


                # This is where the problem is 

                if POCra >= subjpoc_ra[0] and POCra <= subjpoc_ra[1] and \
                POCdec >= subjpoc_dec[0] and POCdec <= subjpoc_dec[1]:
                #if gal_membs['u'][memb] <= -40.71+100 <= 15 and gal_membs['u'][memb] >= -40.71-100 and\
               # gal_membs['v'][memb] <= -19.19+100 and gal_membs['v'][memb] >= -19.19-100 and\
                #gal_membs['w'][memb] <= -0.83+100 and gal_membs['w'][memb] >= -0.83-100:
                
                    raa = []
                    decc = []
                    for i in range(len(members)):
                        ra = []
                        dec = []
                        for x in range(1,5000,10):
                            ra.append(ra0[i]+(pmra[i]*x*1000))
                            dec.append(dec0[i]+(pmdec[i]*x*1000))
                        raa.append(ra)
                        decc.append(dec)

                    # D = [99,96] x [9,6]
                    for i in range(len(raa)):
                        for x in range(len(raa[i])):
                            if raa[i][x] >= POC[0] and raa[i][x] <= POC[1] and \
                            decc[i][x] >= POC[2] and decc[i][x] <= POC[3]:
                                master.append(list(members.loc[i]))
                                break

master = pd.DataFrame(master,columns = list(FILT))
```
