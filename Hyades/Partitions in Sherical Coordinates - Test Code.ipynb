```python
%matplotlib notebook

import math
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from mpl_toolkits import mplot3d

def spherical(x,y,z):
    rho = np.sqrt(x**2+y**2+z**2)
    phi = np.arccos(z/rho)
    theta = np.arctan(y/x)
    lis = [rho,abs(theta),abs(phi)]
    return lis

fig = plt.figure(figsize = (10,10))
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title('Partitions')

RHO, THETA, PHI = list(np.arange(0,300,50)), list(np.mgrid[0:2*np.pi:20j]), list(np.mgrid[0:np.pi:10j])
THETA.append(2*np.pi), PHI.append(np.pi)


lis = [[100,-200,50]]
counter = []

partitions = {} 
dV = 0
for r in range(len(RHO)-1):
    quarter = 0
    for t in range(len(THETA)-1):
        for p in range(len(PHI)-1): 
            ax.scatter3D(RHO[r]*np.cos(THETA[t])*np.sin(PHI[p]),\
                         RHO[r]*np.sin(THETA[t])*np.sin(PHI[p]),\
                         RHO[r]*np.cos(PHI[p]), c='k',s=7)
            #print('---- {} < rho < {} | {} < theta < {} | {} < phi < {}'.format(RHO[r],RHO[r+1],THETA[t],THETA[t+1],PHI[p],PHI[p+1]))
            for i in range(len(lis)):
                coords = spherical(lis[i][0],lis[i][1],lis[i][2])
                if coords[0] >= RHO[r] and coords[1] >= THETA[t] and coords[2] >= PHI[p] and \
                coords[0] <=  RHO[r+1] and coords[1] <= THETA[t+1] and coords[2] <= PHI[p+1]:
                    ax.scatter3D(lis[i][0],lis[i][1],lis[i][2], c='r',s=50)
                     
            partitions[dV] = [[[RHO[r],RHO[r+1]],[THETA[t],THETA[t+1]],[PHI[p],PHI[p+1]]]
            dV += 1
    counter.append(quarter)
            
plt.show()
```
